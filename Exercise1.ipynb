## D. Enriching

Create at least 3 new variables:
- price_per_minimum_night
- review_rate_per_month (combine ratings + frequency)
- is_expensive (boolean flag based on price threshold)


```python
# put your answer here
df['price_per_minimum_night'] = df['price'] / df['minimum_nights']
df['review_rate_per_month'] = df['number_of_reviews'] * df['reviews_per_month']
df['is_expensive'] = df['price'] > 200
display(df.head())
```

Create 1 aggregated summary table

(e.g. average price per neighborhood, number of hosts per borough)

```python
# put your answer here
aggregated_data = df.groupby(['neighbourhood_group', 'room_type']).agg(
    average_price=('price', 'mean'),
    average_reviews=('number_of_reviews', 'mean'),
    total_host_listings=('calculated_host_listings_count', 'sum')
).reset_index()

display(aggregated_data)
```

## E. Validating

Create at least 3 validation checks
- Price must be positive.
- Host ID should not be null.
- Minimum nights must be ≥ 1.
- Generate a validation report:
- Count of rows violating each rule.

Decide if invalid rows (if any) should be removed or corrected?

```python
# put your answer here
# Validation Check 1:
price_violation_count = df[df['price'] <= 0].shape[0]

# Validation Check 2:
host_id_violation_count = df['host_id'].isnull().sum()

# Validation Check 3:
minimum_nights_violation_count = df[df['minimum_nights'] < 1].shape[0]

# validation report
validation_report = pd.DataFrame({
    'Validation Rule': [
        'Price must be positive',
        'Host ID should not be null',
        'Minimum nights must be \u2265 1'
    ],
    'Violating Rows Count': [
        price_violation_count,
        host_id_violation_count,
        minimum_nights_violation_count
    ]
})

display(validation_report)

# Decide if invalid rows (if any) should be removed or corrected?
if price_violation_count == 0 and host_id_violation_count == 0 and minimum_nights_violation_count == 0:
    print("\nDecision on invalid rows:")
    print("Based on the cleaning steps performed earlier:")
    print(" - Rows with `price <= 0` were already removed.")
    print(" - Rows with missing `host_name` (and implicitly `host_id`) were removed through `dropna()`.")
    print(" - `minimum_nights` are all \u2265 1 based on previous descriptive analysis.")
    print("Therefore, at this stage, no further rows need to be removed or corrected based on these specific rules, as the dataset already adheres to them.")
else:
    print("\nDecision on invalid rows: Some violations found. Review the report above for details and consider removal or correction based on the context.")
```
